```python
import os

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
archived_runs = [run for run in os.listdir(os.getcwd()) if run.startswith('Run')]
archived_runs.sort()
results = {
    run:
        {
            "final_test": pd.read_csv(os.path.join(os.getcwd(), run, 'final_test.csv')),
            "training_loss": pd.read_csv(os.path.join(os.getcwd(), run, 'training_loss.csv')),
            "validation_results":  pd.read_csv(os.path.join(os.getcwd(), run, 'validation_results.csv'))
        }
        for run in archived_runs
}
```

```python
fig, axes = plt.subplots(nrows = len(archived_runs), ncols=1, figsize=(10, 15))
for ax, (run_name, run_results) in zip(axes, results.items()):
    ax.plot(run_results['training_loss']['loss'], label='training loss')
    ax.plot(run_results['validation_results']['test_loss'], label='validation loss')
    ax.set_title(run_name)
    ax.legend()
fig.tight_layout()
```

```python
fig, axes = plt.subplots(nrows = len(archived_runs), ncols=1, figsize=(10, 15))
for ax, (run_name, run_results) in zip(axes, results.items()):
    ax.plot(run_results['validation_results']['accuracy'], label='accuracy')
    ax.set_title(run_name)
    ax.legend()
fig.tight_layout()
```

```python
fig, axes = plt.subplots(nrows = len(archived_runs), ncols=1, figsize=(10, 15))
for ax, (run_name, run_results) in zip(axes, results.items()):
    ax.plot(run_results['validation_results']['accuracy'], label='accuracy')
    ax.plot(run_results['validation_results']['f1score'], label='f1score')
    ax.plot(run_results['validation_results']['precision'], label='precision')
    ax.plot(run_results['validation_results']['recall'], label='recall')
    ax.set_title(run_name)
    ax.legend()
fig.tight_layout()
```

```python
fig, axes = plt.subplots(nrows = len(archived_runs), ncols=1, figsize=(10, 15))
for ax, (run_name, run_results) in zip(axes, results.items()):
    ax.plot(run_results['training_loss']['gradients_norm'], label='gradients_norm')
    ax.plot(run_results['training_loss']['weights_norm'], label='weights_norm')
    ax.set_title(run_name)
    ax.legend()
fig.tight_layout()
```
