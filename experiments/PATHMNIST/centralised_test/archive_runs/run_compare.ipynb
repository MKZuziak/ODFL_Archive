```python
import os

import pandas as pd
import matplotlib.pyplot as plt
```

```python
RUNS = ['Run_III', 'Run_IV', 'Run_V', 'Run_VI']
METRICS = ['accuracy', 'f1score', 'precision', 'recall']
fig, axes = plt.subplots(len(RUNS), 1, figsize=(20, 15))
for ax_num, metric in enumerate(METRICS):
    axes[ax_num].set_title(metric)
    for run in RUNS:
        validation_results = pd.read_csv(os.path.join(os.getcwd(), run, 'validation_results.csv'))
        axes[ax_num].plot(validation_results[metric], label=run)
    axes[ax_num].legend()
fig.tight_layout()
```
